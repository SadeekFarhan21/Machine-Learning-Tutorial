```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv')
le = LabelEncoder()
df['company'] = le.fit_transform(df['company'])
df['job'] = le.fit_transform(df['job'])
df['degree'] = le.fit_transform(df['degree'])
target = df['salary_more_then_100k']
feature = df[df.columns.drop(['salary_more_then_100k'])]
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.1)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
sns.heatmap(confusion_matrix(y_test, model.predict(X_test)), annot = True)
```
